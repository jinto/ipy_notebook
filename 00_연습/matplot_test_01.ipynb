```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np


vals = [3,2,5,0,1]
plt.plot(vals)
```

```python
x = np.arange(0, 20, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

x1,x2,y1,y2 = plt.axis()
plt.axis((x1,20,-1,1))

plt.plot(x, y_sin)
plt.plot(x, y_cos)
```

```python

counts =  {
    'Snow Belle': 63,
    'Champion': 76,
    'Cherry Belle': 58,
    'Daikon': 63,
    'French Breakfast': 72,
    'Bunny Tail': 72,
    'Sicily Giant': 57,
    'Red King': 56,
    'Plum Purple': 56,
    'April Cross': 72
}
names = []
votes = []
# Split the dictionary of name:votes into two lists, one for names and one for vote count
for radish in counts:
    names.append(radish)
    votes.append(counts[radish])

# The X axis can just be numbered 0,1,2,3...
x = np.arange(len(counts))

plt.bar(x, votes)
plt.xticks(x + 0.5, names, rotation=90)
```
